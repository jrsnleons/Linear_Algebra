- To complete the process of determining the solution, create a function or steps of function that will automate backward substitution.


```python
import numpy as np

def BackwardSubstitution(A):
    """
    Perform backward substitution on the row-reduced echelon form of a matrix A
    to find the solution to the system of linear equations.

    Parameters:
    A (numpy.ndarray): Row-reduced echelon form of the augmented matrix.

    Returns:
    numpy.ndarray: Solution vector.
    """
    m, n = A.shape
    x = np.zeros(m)

    for i in range(m - 1, -1, -1):
        x[i] = A[i, -1] / A[i, i]
        for j in range(i - 1, -1, -1):
            A[j, -1] -= A[j, i] * x[i]

    return x

# Example Usage:
# Assuming A is the row-reduced echelon form of the augmented matrix
A = np.array([[1, -1, 1, 3],
              [2, 1, 8, 18],
              [4, 2, -3, -2]])
Solution = BackwardSubstitution(A)
print(Solution)
```
